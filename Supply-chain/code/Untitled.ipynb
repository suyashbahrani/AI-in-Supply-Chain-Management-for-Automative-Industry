```python
from tkinter import *
from tkinter.filedialog import askopenfilename
import numpy as np
import cv2

root = Tk()
root.title("IMAGE PROCESSING ")
root.geometry("450x200+200+100")

file_name = " "
filter = " "

mylabel = Label(root, text='Choose image')
mylabel.grid(row=12, column=12, padx=20, pady=20)

b1 = Button(text="Open")
b1.grid(row=12, column=14, padx=20, pady=20)

b2 = Button(text="Enter")
b2.grid(row=28, column=12, padx=20, pady=20)

e1 = Entry(root,bg="white")
e1.grid(row=12, column=13)

tkvar = StringVar(root)

choices = {'Invert','Edge Detection','Blur','Smooth','Crop'}
tkvar.set('Invert')

def dropdown(event):
    global filter
    filter = tkvar.get()
    print(filter)

popupMenu = OptionMenu(root, tkvar, *choices, command=dropdown)
Label(root, text="Choose filter").grid(row=20, column=12)
popupMenu.grid(row=20, column=13)

root.configure(background='black')

def onclick(event):
    print("You clicked the button")
    Tk().withdraw()
    filename = askopenfilename()
    global file_name
    file_name=filename
    e1.insert(0, file_name)
    print(file_name)

def onclick1(event):
    global img
    img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)
    print("You clicked the enter button")
    print(file_name)
    if filter == 'Invert':
        from matplotlib import pyplot as plt
        img1= cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)
        height = img1.shape[0]
        width = img1.shape[1]
        max_intensity = 255
        for i in np.arange(height):
            for j in np.arange(width):
                a = img1.item(i, j)
                b = max_intensity - a
                img1.itemset((i, j), b)
        plt.subplot(121), plt.imshow(img, cmap='gray')
        plt.title('Original Image'), plt.xticks([]), plt.yticks([])
        plt.subplot(122), plt.imshow(img1, cmap='gray')
        plt.title('Inverted Image'), plt.xticks([]), plt.yticks([])
        plt.show()
    elif filter == 'Edge Detection':
        from matplotlib import pyplot as plt
        edges = cv2.Canny(img, 100, 200)
        plt.subplot(121), plt.imshow(img, cmap='gray')
        plt.title('Original Image'), plt.xticks([]), plt.yticks([])
        plt.subplot(122), plt.imshow(edges, cmap='gray')
        plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
        plt.show()
    elif filter == 'Blur':
        from matplotlib import pyplot as plt
        blur = cv2.blur(img, (100, 100))
        plt.subplot(121), plt.imshow(img, cmap='gray')
        plt.title('Original Image'), plt.xticks([]), plt.yticks([])
        plt.subplot(122), plt.imshow(blur, cmap='gray')
        plt.title('Blurred Image'), plt.xticks([]), plt.yticks([])
        plt.show()
    elif filter == 'Smooth':
        from matplotlib import pyplot as plt
        mask = np.ones((3, 3), np.float32) / 1
        dst = cv2.filter2D(img, -1, mask)
        plt.subplot(121), plt.imshow(img, cmap='gray')
        plt.title('Original Image'), plt.xticks([]), plt.yticks([])
        plt.subplot(122), plt.imshow(dst, cmap='gray')
        plt.title('Smooth Image'), plt.xticks([]), plt.yticks([])
        plt.show()
    elif filter == 'Crop':
        from matplotlib import pyplot as plt
        print(img.shape)
        crop = img[100:200, 100:200]
        print(crop)
        plt.subplot(121), plt.imshow(img, cmap='gray')
        plt.title('Original Image'), plt.xticks([]), plt.yticks([])
        plt.subplot(122), plt.imshow(crop, cmap='gray')
        plt.title('Cropped Image'), plt.xticks([]), plt.yticks([])
        plt.show()
    Tk().withdraw()

b1.bind('<Button-1>', onclick)
b2.bind('<Button-1>', onclick1)

mainloop()
```
